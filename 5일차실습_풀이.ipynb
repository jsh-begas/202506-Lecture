```python
import pandas as pd
```

```python
main_df = pd.read_excel('data/car_data_2011_main.xlsx')
```

```python
car_type_df = pd.read_csv('data/car_type.csv')
```

```python
main_df
```

```python
car_type_df
```

```python
merge_df = pd.merge(
    main_df
    ,car_type_df
    ,how='left'
    ,on='Car_Name'
)
```

```python
transmission_flag = merge_df['Transmission']=='Manual'
```

```python
price_flag = merge_df['Present_Price']>=3
```

```python
merge_df[transmission_flag & price_flag]
```

```python
merge_df.pivot_table(
    index = 'Seller_Type'
    ,values = 'Selling_Price'
    ,aggfunc = 'sum'
).to_csv('selling_price_sum.csv')
```
