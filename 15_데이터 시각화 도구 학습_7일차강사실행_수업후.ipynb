```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.get_backend()
```

```python
plt.rc('font', family = 'NanumGothic')       # 한글 깨짐 해결
mpl.rcParams['axes.unicode_minus'] = False   # plot 마이너스가 깨질 경우

#그림을 plt.show()나 fig 등으로 명시적으로 출력하지 않아도 셀 안에서 그림을 생성한 경우 출력됨
%matplotlib inline
```

```python
# 빈 figure 정의
fig = plt.figure()

# figure에 axes 추가
# figure 전체가 [0,1] x [0,1]이라 치고 화면에 그림을 그릴 영역인 사각형을 추가
# rect : [x축의 시작 위치, y축의 시작 위치, 이미지의 가로 길이, 이미지의 세로 길이]
ax1 = fig.add_axes([0  ,0  ,0.8,0.8])
ax2 = fig.add_axes([0.5,0.5,0.5,0.5])
```

```python
# 만일 fig에서 다시 ax 객체를 꺼내고 싶다면
ax1_again = fig.axes[0]
ax2_again = fig.axes[1]
print(ax1 == ax1_again)
print(ax2 == ax2_again)
```

```python
# 위쪽 fig에 내용을 추가해서 '다시' 그리기
ax1.plot([1,2,3],[4,5,6])
fig
```

```python
# plt.subplots 형식으로 fig, fig내 axes를 생성하고 창을 나누기
fig, axs = plt.subplots(1, 3, squeeze=False)
axs[0,1].plot([1,2,3],[4,5,6])
```

```python
import pandas as pd
import numpy as np
```

```python
# 예제 데이터 생성
plot_df = pd.DataFrame({
    'x': np.linspace(0, 10, 10)
    ,'y': np.linspace(0, 50, 10)
})
plot_df
```

```python
# 편하지만, 비추천
plt.plot(plot_df['x'],plot_df['y'])
```

```python
fig, ax = plt.subplots()
ax.plot(plot_df['x'],plot_df['y'])
ax.set_title('선 도표')
ax.set_xlabel('x축 이름')
ax.set_ylabel('y축 이름')
```

```python
fig, ax = plt.subplots()
ax.scatter(plot_df['x'],plot_df['y'])
ax.set_title('산점도')
```

```python
# 막대 도표
fig, ax = plt.subplots()
ax.bar(plot_df['x'],plot_df['y'])
ax.set_title('막대 도표')
```

```python
fig, ax = plt.subplots()
ax.barh(plot_df['x'],plot_df['y'])
ax.set_title('막대 도표')
```

```python
# 겹쳐 그리기
fig, ax = plt.subplots()
ax.bar(plot_df['x'],plot_df['y'])
ax.scatter(plot_df['x'],plot_df['y'], color='red')
ax.set_title('막대 도표+점 도표')
```

```python
# 히스토그램 = 값 세기 + 막대그래프
fig, ax = plt.subplots()
ax.hist(plot_df['x'], edgecolor='black', bins=3)
```

```python
fig, ax = plt.subplots()
ax.hist2d(plot_df['x'],plot_df['y'])
```

```python
import seaborn as sns
iris_df = pd.read_csv('data/iris_data.csv')
```

```python
iris_df
```

```python
fig, ax = plt.subplots()
sns.histplot(
    data = iris_df
    ,x = 'sepal_length'
    ,ax = ax
)
```

```python
fig, ax = plt.subplots()
sns.boxplot(
    data = iris_df
    ,x = 'species'
    ,y = 'sepal_length'
    ,ax = ax
)
```

```python
fig, ax = plt.subplots()
sns.kdeplot(
    data = iris_df
    ,x = 'sepal_length'
    ,hue = 'species'
    ,ax = ax
    ,bw_adjust=1
)
ax.scatter(iris_df['sepal_length'],np.zeros(iris_df.shape[0]))
ax.set_title('KDE(밀도)')
```

```python
ax.scatter
```

```python
fig, ax = plt.subplots()
sns.scatterplot(
    data = iris_df
    ,x = 'sepal_length'
    ,y = 'sepal_width'
    ,hue = 'species'
    ,size = 'petal_length'
    ,ax = ax
)
```

```python
fig, ax = plt.subplots()
ax.hist2d(
    iris_df['sepal_length'], iris_df['sepal_width']
    ,bins = [20,15]
)
```

```python
fig, ax = plt.subplots()
sns.histplot(
    data = iris_df
    ,x = 'sepal_length'
    ,y = 'sepal_width'
    ,hue = 'species'
    ,binwidth = [0.2,0.2]
)
```

```python
fig, ax = plt.subplots()
sns.kdeplot(
    data = iris_df
    ,x = 'sepal_length'
    ,y = 'sepal_width'
    ,hue = 'species'
    ,fill = True
)
```
